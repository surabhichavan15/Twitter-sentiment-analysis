# Text cleaning

```python
stop_words = set(stopwords.words("english"))

def cleaner(data):
    
    # Removing digits
    digit_removed=str(data).replace("\d","")# removing digits
    
    # Tokens
    tokens = word_tokenize(str(digit_removed).replace("'", "").lower()) 
    
    # Remove Puncs
    without_punc = [w for w in tokens if w.isalpha()]
    
    # Stopwords
    without_sw = [t for t in without_punc if t not in stop_words]
    
    # Lemmatize
    text_cleaned = [WordNetLemmatizer().lemmatize(t) for t in without_sw]
 
    return " ".join(text_cleaned)

```

```python
train_df["Text"] = train_df["Text"].apply(cleaner)
train_df.head()
```

```python
def cleaner1(data):
    
    # Removing digits
    digit_removed1=str(data).replace("\d","")# removing digits
    
    # Tokens
    tokens1 = word_tokenize(str(digit_removed1).replace("'", "").lower()) 
    
    # Remove Puncs
    without_punc1 = [w for w in tokens1 if w.isalpha()]
    
    # Stopwords
    without_sw1 = [t for t in without_punc1 if t not in stop_words]
    
    # Lemmatize
    text_cleaned1 = [WordNetLemmatizer().lemmatize(t) for t in without_sw1]
    
    return " ".join(text_cleaned1)
```

```python
test_df["Text"] = test_df["Text"].apply(cleaner1)
test_df.head()
```
